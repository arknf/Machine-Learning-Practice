```python
#splitting features and target
X = train.drop(['price_range'],axis=1)
y = train['price_range']
```

```python
#selecting features
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

select_feature = SelectKBest(score_func=chi2,k=20)
ordered_feature = select_feature.fit(X,y)

dfscores = pd.DataFrame(ordered_feature.scores_,columns=["Score"])
dfcolumns = pd.DataFrame(X.columns)

features_rank=pd.concat([dfcolumns,dfscores],axis=1)
features_rank.columns=['Features','Score']
features_rank.sort_values(by='Score',ascending=False).reset_index().drop(['index'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

#using train_test_split
X_train , X_val , y_train , y_val = train_test_split(X, y, test_size = 0.25, random_state = 10)
```

```python
#Importing Classifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.metrics import mean_squared_error, accuracy_score
def count_rmse_and_score(classifier):
    model = classifier()
    model.fit(X_train,y_train)
    pred = model.predict(X_val)
    RMSE = mean_squared_error(y_val,pred)
    SCORE = accuracy_score(y_val,pred)
    return RMSE,SCORE
```

```python
print(count_rmse_and_score(KNeighborsClassifier)),
print(count_rmse_and_score(DecisionTreeClassifier)),
print(count_rmse_and_score(RandomForestClassifier)),
print(count_rmse_and_score(XGBClassifier)),
print(count_rmse_and_score(SVC)),
print(count_rmse_and_score(GaussianNB))
```

```python
score_list = [count_rmse_and_score(KNeighborsClassifier)[1],
             count_rmse_and_score(DecisionTreeClassifier)[1],
             count_rmse_and_score(RandomForestClassifier)[1],
             count_rmse_and_score(XGBClassifier)[1],
             count_rmse_and_score(SVC)[1],
             count_rmse_and_score(GaussianNB)[1]]
classifier_list = ['KNeighborsClassifier',
                  'DecisionTreeClassifier',
                  'RandomForestClassifier',
                  'XGBClassifier',
                  'SVC',
                  'GaussianNB']
plt.scatter(classifier_list,score_list)
plt.xlabel('classifier')
plt.ylabel('score')
plt.xticks(rotation=90)
```

```python
#tuning KNeighborsClassifier
from sklearn.model_selection import RandomizedSearchCV

#parameter
leaf_size = list(range(1,100))
n_neighbors = list(range(1,50))
p=[1,2]

#hyperparameter
hyperparameters = dict(leaf_size=leaf_size, n_neighbors=n_neighbors, p=p)

#Use GridSearch
knn = KNeighborsClassifier()
clf = RandomizedSearchCV(knn, hyperparameters, cv=10)

#Fit the model
knn_model = clf.fit(X_train,y_train)
```

```python
knn_model.best_params_
```

```python
kn_new = KNeighborsClassifier(p = 1, n_neighbors = 25, leaf_size = 89)
kn_new.fit(X_train,y_train)
kn_pred = kn_new.predict(X_val)
kn_acc = accuracy_score(y_val,kn_pred)
kn_acc
```

```python
from sklearn.model_selection import StratifiedKFold

clf = SVC()
params = {'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
          'degree': np.linspace(2, 5, 4),
          'C': np.logspace(-3, 5, 17),
          'gamma': np.logspace(-3, 5, 17)}

random_search = RandomizedSearchCV(
    estimator=clf, param_distributions=params, n_iter=200, n_jobs=-1,
    cv=StratifiedKFold(n_splits=5))

random_search.fit(X_train,y_train)
```

```python
random_search.best_params_
```

```python
svc_new = SVC(kernel = 'linear',
              gamma = 31622.776601683792,
              degree = 3.0,
              C = 0.001)
svc_new.fit(X_train,y_train)
svc_pred = svc_new.predict(X_val)
svc_acc = accuracy_score(y_val,svc_pred)
svc_acc
```
