```python
#function to find Missing Value in every columns
def get_missing_value(df):
    mis_val = df.isnull().sum()
    tot_val = df.shape[0]
    mis_percentage = round((100 * mis_val / tot_val),2)
    miss_tab = pd.concat([mis_val, mis_percentage], axis=1)
    miss_tab_col_name = miss_tab.rename(columns = {0 : 'Row With Missing Value', 1 : 'Percentage (%)'})
    miss_tab_col_name = miss_tab_col_name.sort_values('Percentage (%)',ascending=False)
    return miss_tab_col_name
```

```python
#checking dataset missing row/values
get_missing_value(housing)
```

```python
housing.dropna(inplace=True)
```

```python
housing = housing.drop(['Address', 'Date'], axis = 1)
```

```python
housing
```

```python
#Label Encoding for categorical features (train)
from sklearn.preprocessing import LabelEncoder
for i in housing[housing.dtypes[housing.dtypes == "object"].index].columns:
    le = LabelEncoder() 
    le.fit(list(housing[i].values)) 
    housing[i] = le.transform(list(housing[i].values))
housing
```

```python
#Modelling
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
#Correlation any features to Price
corr_0 = housing.corr()
corr_0.sort_values(["Price"], ascending = False, inplace = True)
np.abs(corr_0.Price[abs(corr_0.Price)>0]).sort_values()
```

```python
features = ['BuildingArea','Rooms','Bathroom','Bedroom2','Type','YearBuilt','Distance','Lattitude','Longtitude','Car']
X = housing.drop(['Price'],axis=1)
y = housing['Price']
```

```python
#using train_test_split
X_train , X_val , y_train , y_val = train_test_split(X, y, test_size = 0.25, random_state = 10)
```

```python
def count_rmse_and_score(regressor):
    model = regressor()
    model.fit(X_train,y_train)
    pred = model.predict(X_val)
    RMSE = mean_absolute_error(y_val,pred)
    return RMSE
```

```python
print(count_rmse_and_score(LinearRegression),
count_rmse_and_score(Lasso),
count_rmse_and_score(Ridge),
count_rmse_and_score(ElasticNet),
count_rmse_and_score(RandomForestRegressor),
count_rmse_and_score(GradientBoostingRegressor),
count_rmse_and_score(DecisionTreeRegressor),
count_rmse_and_score(KNeighborsRegressor))
```

```python
#Hyperparameter Tuning
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
rf_pred = rf_random.predict(X_val)
rf_RMSE = mean_absolute_error(y_val,rf_pred)
rf_RMSE
```
